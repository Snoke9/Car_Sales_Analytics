###6. Проверьте датафрейм на наличие отсутствующих значений и дубликатов строк. При наличии обработайте их тем или иным способом

```python
# проверка на наличие дубликатов строк
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# проверка на пропущенные значения
print(df.isnull().sum())
```

```python
# заполним пропуски в столбце four_year_resale_value, вычислив средний арифметический коэффициент, на который падает стоимость
mask = df['four_year_resale_value'].notnull() & df['Price_in_thousands'].notnull()
k = (df.loc[mask, 'four_year_resale_value'] / df.loc[mask, 'Price_in_thousands']).mean()
df['four_year_resale_value'] = df['four_year_resale_value'].fillna(df['Price_in_thousands'] * k)
# убедимся что пропусков не осталось
df['four_year_resale_value'].isna().sum()
```

###9. Найдите десять моделей, больше всего потерявших в цене за четыре года и десять моделей, меньше всего потерявших в цене

```python
df_price_lost = df.copy()
df_price_lost['price_lost'] = df_price_lost['Price_in_thousands'] - df_price_lost['four_year_resale_value']
```

####10 моделей с наибольшей потерей в цене

```python
df_price_lost.nlargest(10, 'price_lost')[['Manufacturer', 'Model', 'Price_in_thousands', 'four_year_resale_value', 'price_lost']]
```

####10 моделей с наименьшей потерей в цене

```python
df_price_lost.nsmallest(10, 'price_lost')[['Manufacturer', 'Model', 'Price_in_thousands', 'four_year_resale_value', 'price_lost']]
```

###12. Вычислите для столбца Price_in_thousands среднее значение, моду, медиану, дисперсию и среднеквадратичное отклонение

```python
mean_Price_in_thousands = df['Price_in_thousands'].mean()
mode_Price_in_thousands = df['Price_in_thousands'].mode()[0]
median_Price_in_thousands = df['Price_in_thousands'].median()
variance_Price_in_thousands = df['Price_in_thousands'].var()
std_dev_Price_in_thousands = df['Price_in_thousands'].std()
print(f'Среднее значение: {mean_Price_in_thousands}\n'
f'Мода: {mode_Price_in_thousands}\n'
f'Медиана: {median_Price_in_thousands}\n'
f'Дисперсия: {variance_Price_in_thousands}\n'
f'Среднеквадратичное отклонение: {std_dev_Price_in_thousands}')
```

###13. Зависит ли цена автомобиля от коэффициента Power_perf_factor?

```python
plt.scatter(df['Power_perf_factor'], df['Price_in_thousands'])
plt.xlabel("Максимальный коэффициент полезного действия")
plt.ylabel("Цена в тыс. долларах")
plt.title("Зависимость цены автомобиля от коэффициента полезного действия")
plt.show()
```

####По графику зависимости видно, что при увеличении коэффициента полезного действия растёт и цена автомобиля.
####Также посчитаем коэффициент корреляции:

```python
print(f'Коэффициент корреляции: {df['Price_in_thousands'].corr(df['Power_perf_factor'])}')
```

####Коэффициент близок к единице — значит существует сильная прямая зависимость.
###14. Найдите лучшие модели по соотношению цена-качество



```python
df_quality = df.copy()
# введем новый столбец "качество"
df_quality['quality'] = (df_quality['Horsepower'] + df_quality['Fuel_efficiency'] + df_quality['Power_perf_factor']) / df_quality['Price_in_thousands']
df_quality.nlargest(10, 'quality')[['Manufacturer', 'Model', 'Price_in_thousands', 'Horsepower', 'Fuel_efficiency', 'Power_perf_factor', 'quality']]
```
